1. Sunday Most Travelled Routes problem.

```python
#Most travelled routes by Travel route frequency
df['travel_from'].value_counts()
```

```python
# We develop a query that fetches data for travels on Sunday only, storing such to a new DataFrame.
df2 = df.query('Day=="Sunday"')
```

```python
#New DataFrame.
df2.head()
```

```python
#We extract the routes by travel frequency to Nairobi and store the data in a dictionary.
sunday_route_frequency_dict = df2['travel_from'].value_counts().to_dict()
```

```python
#extracted dictionary with our interest data
sunday_route_frequency_dict
```

```python
#We iterate over the dictionary and print the travel frequencies count
for route in sunday_route_frequency_dict:
    print(sunday_route_frequency_dict[route])
```

```python
#We convert the dictionary to a list and further to a dataFrame for easy visualization.
sunday_dataframe = pd.DataFrame(list(sunday_route_frequency_dict.items()),columns=['route','count'])
```

```python
sunday_dataframe.head() 
```

```python
#We plot a graph to that we can map the top 7 most travelled routes by frequency of root travel.
sunday_dataframe.iloc[0:7].plot(kind='bar',x='route',y='count',xlabel='Travel route',ylabel='Frequency',title='The most travel from routes to Nairobi on a sunday')
```

3. Payment Reference character sequence problem.

```python
#The for loop finds the occurence of the sequence MK in the payment refrence data
#we store the refrence to a new array -reciept
count = 0
receipt = []
for k in df['payment_receipt']:
    resu = 'MK' in k
    if resu==True:
        receipt.append(k)
        count+=1
    
    
```

```python
receipt
```

```python
#We iterate over the payment refrence to find the next character after MK,
#if it is the last character we return a "no value".
characters=[]

for a in receipt:
    pos = a.find('MK')
    if pos<8:
        characters.append(a[pos+2])
    else:
        characters.append('no character')
```

```python
characters_after_MK_in_receipt_sequence = pd.DataFrame(characters,columns=['count'])
```

```python
#the character and their occurence frequency
characters_after_MK_in_receipt_sequence['count'].value_counts()
```

```python
#Probability of Non Occurrence.
52/len(characters_after_MK_in_receipt_sequence)
```
